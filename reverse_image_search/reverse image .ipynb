### What is ResNets



ResNet can have a very deep network of up to 152 layers by learning the residual representation functions instead of learning the signal representation directly.

ResNet introduces skip connection (or shortcut connection) to fit the input from the previous layer to the next layer without any modification of the input. Skip connection enables to have deeper network and finally ResNet becomes the Winner of ILSVRC 2015 in image classification, detection, and localization, as well as Winner of MS COCO 2015 detection, and segmentation.

Now we are taken the image has [base64](https://en.wikipedia.org/wiki/Base64) decoder Base64 encoding tool supports loading the [Base64](https://stackabuse.com/encoding-and-decoding-base64-strings-in-python/) text File to transform to Image 'Click on the Upload File button and select File'.

```python
def predict(file):
    x = image=Image.open(file)
    img_t = transform(image)
    image = torch.unsqueeze(img_t, 0)
    # Carry out model inference
    out = model(image.to(device))
    # Forth, print the top 5 classes predicted by the model
    _, indices = torch.sort(out, descending=True)
    
    return mappings[indices[0][0].item()]
```

this the predict function where we upload the file in upload folder and gives the output finally returns the result. 

```python
def my_random_string(string_length=10):
    """Returns a random string of length string_length."""
    random = str(uuid.uuid4()) # Convert UUID format to a Python string.
    random = random.upper() # Make all characters uppercase.
    random = random.replace("-","") # Remove the UUID '-'.
    return random[0:string_length] # Return the random string.

def allowed_file(filename):
    return '.' in filename and \
           filename.rsplit('.', 1)[1] in ALLOWED_EXTENSIONS
```

these are the helping functions that are going to save the image in the upload folder then its going to check the allowed file extensions
